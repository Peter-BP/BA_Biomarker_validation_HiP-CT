### Convert from jp2 to png for Root Painter annotation

Root painter trains on png files and the Human Organ Atlas files are jp2. This converts for training.

```python
import os
import numpy as np
from PIL import Image

source_folder = 'C:/Users/Peter/Desktop/root_painter/sets/BA training - 100.0um_LADAF-2021-17_kidney-right_complete-organ_pag-0.02_0.04_jp2_'  
output_folder = 'C:/Users/Peter/Desktop/root_painter/sets/BA training_png' 

if not os.path.exists(output_folder):
    os.makedirs(output_folder)

for filename in os.listdir(source_folder):
    if filename.lower().endswith('.jp2'):
        img_path = os.path.join(source_folder, filename)
        img = Image.open(img_path)
            
        if img.mode in ('I', 'I;16', 'F'):
            arr = np.array(img)

            if arr.max() > arr.min():
                arr = (arr - arr.min()) / (arr.max() - arr.min()) * 255.0
                
            img = Image.fromarray(arr.astype('uint8'))

        img = img.convert("RGB")
            
        new_filename = os.path.splitext(filename)[0] + ".png"
        output_path = os.path.join(output_folder, new_filename)
            
        img.save(output_path, 'PNG', compress_level=6)
            
        print(f"Converted to png: {filename}")

print("Done.")
```

### Convert from .png segmentation from root painter to .tif files for pipeline

My pipeline tages segmentation done in .tif, this convertes .png segmentations from root painter to .tif.

```python
import os
from PIL import Image

input_folder = 'C:/Users/Peter/Desktop/Bachelorprojekt/segmentation done/BA test 2' 
output_folder = 'C:/Users/Peter/Desktop/Bachelorprojekt/segmentation done/BA test 2_conv'   

if not os.path.exists(output_folder):
    os.makedirs(output_folder)

for filename in os.listdir(input_folder):
    if filename.lower().endswith('.png'):
        img_path = os.path.join(input_folder, filename)
            
        img = Image.open(img_path)
            
        new_filename = os.path.splitext(filename)[0] + ".tif"
        output_path = os.path.join(output_folder, new_filename)
            
        img.save(output_path, format='TIFF', compression='None')
            
        print(f"{new_filename} Converted to .tif" )

print("Conversion complete.")
```
