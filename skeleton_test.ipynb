### For visualizating a vessel segmentation and it's strahler tree

PATH_TO_FILE is the directory to the segmentation

```python
from pyth.dataload import load_kidney_labels_robust
PATH_TO_FILE = "C:/Users/Peter/Desktop/Bachelorprojekt/Data/blood-vessel-segmentation/train/kidney_2/labels"
voxel_grid = load_kidney_labels_robust(PATH_TO_FILE, scale=0.5, use_cache = True)
```

```python
from pyth.skelet import skeletonize_voxel_grid
network, dist_map = skeletonize_voxel_grid(voxel_grid)
```

```python
from pyth.TVis import visualize_voxel_tree
visualize_voxel_tree(voxel_grid,bounds_min=[-128, -128, -128], voxel_size=1, realdata=True)
```

```python
from pyth.TVis import plot_strahler_3d
plot_strahler_3d(network)
```

```python
from pyth.TVis import plot_vessel_network
plot_vessel_network(network)
```
